## 이동평균선이란?

- 평균이라는 개념을 가지고 차트를 분석하는데 사용
- 지지선과 저항선의 역할을 수행

- 가장 최근 일주일치 종가를 평균내면 = 이동평균(5일 선)
  - 주로 5일, 20일, 60일, 120일 사용

이동평균과 평균의 가장 큰 차이점은 <b>시간</b>이라는 개념인데,
평균은 동일시점에서 산출되는 것이 흔한 반면, 이동평균은 동일 대상이지만 시점이 서로 상이해서 발생한다.

2002년 16강 월드컵 때 패널티 킥을 실축한 시점에서 안정환과 연장전에서 역전 결승골을 넣은 안정환은 동일 인물이지만, 불과 몇시간 만에 사람들에게서 평가가 전혀 달라진다. 


```
1. 패널티 킥 진전의 안정환

2. 패널티 킥 실축 후 안정환

3. 역전 골 후
 안정환
방법이 존재
```


팬들에 대한 기대치를 산출할 때, 각 시점의 기대치를 단순하게 합해서 나누는 방법을 사용할 수도 있지만, 가장 최근 기대치에 가장 많은 가중치를 주어서 측정할 수도 있다. 또한 가장 마지막 기대치를 통해 측정할 수도 있다. 


그 결과 이동평균을 구하는 방법엔


단순이동평균(Simple Moving Average)  
- 일반적인 평균을 구하는 방법

가중이동평균(Weighted Moving Average) : 최근의 데이터에 더 많은 가중치를 부여
- n1 , n2, n3 가 존재한다고 가정하고 각각의 가중치를 w1, w2, w3라고 하면

(w1n1 + w2n2 + w3*n3) / (w1 + w2 + w3
) 

지수이동평균(Exponential Moving Average: 가중치가 지수적으로 감소하는 방식을 사용하여 이동평균 계산 

- 이동평균선을 사용하면 데이터의 장기적인 추세를 보다 명확하게 볼 수 있으며, 단기적인 변동이나 노이즈를 줄일 수 있다.존재

## 사용 설명서

- 가격 < 이동평균 가격
    - 해당 기간 동안 매수에 임했던 사람들이 현재 손실을 보고 있다고 해석
    - 가격이 이동평균선까지 상승하게 된다면 본전을 찾고자 하는 매도세가 생김
- 가격 > 이동평균 가격
    - 해당 기간동안 매수에 임했던 사람들이 현재 수익을 보고 있다고 해석
    - 가격이 이동평균선까지 하락하게 된다면 저점 매수에 대한 희망으로 추가적인 매수세가 생길 수 있음

````python
import FinanceDataReader as fdr
from datetime import datetime
import pandas as pd

import matplotlib.pyplot as plt
````

````python
today = datetime.today().strftime('%Y-%m-%d')

df = fdr.DataReader("005930", start="2022-1-1", end=today)
````

````python
df['MA5']=df['Close'].rolling(5).mean()
df['MA20']=df['Close'].rolling(20).mean()
df['MA60']=df['Close'].rolling(60).mean()
df['MA120']=df['Close'].rolling(120).mean()
````

````python
df.tail(4)
````

````python
plt.figure(figsize=(14, 10))
plt.plot(df.index, df["Close"], label='close', color='gray')
plt.plot(df.index, df["MA5"], label='5-day',color='red')
plt.legend()
plt.show()
````

- 5일선은 최근 5일 동안의 평균 매매 가격으로 단기 매매선이라고 부른다.
- 5일선의 기울기는 대부분 현재의 추게를 그대로 반영
- 기울기를 통해 현재 단기적으로 상승 추세인지, 하락 추세인지를 판단할 수 있음

````python
plt.figure(figsize=(14, 10))
plt.plot(df.index, df["Close"], label='close', color='gray')
plt.plot(df.index, df["MA20"], label='20-day',color='blue')
plt.legend()
plt.show()
````

- 20일 선은 중기 매매선, 심리선 혹은 생명선이라고 부른다.
- 일봉 차트에서 차지하는 비중이 매우 높음. 추세가 살아있을 때 주요 지지선의 역할을 수행한다

````python
plt.figure(figsize=(14, 10))
plt.plot(df.index, df["Close"], label='close', color='gray')
plt.plot(df.index, df["MA5"], label='5-day',color='red')
plt.plot(df.index, df["MA20"], label='20-day',color='blue')
plt.plot(df.index, df["MA60"], label='60-day',color='pink')
plt.legend()
plt.show()
````

- 60일 선은 중기적 추세선 혹은 수급선이라고 불린다. 만약 5일선, 20일선, 60일선 차례로 배열된 상태를 차트의 정배열 상태라고 하는데, 이 상태에서 60일 선이 주요 지지선의 역할을 하는 경우가 많음

````python
plt.figure(figsize=(14, 10))
plt.plot(df.index, df["Close"], label='close', color='gray')
plt.plot(df.index, df["MA120"], label='120-day',color='black')
plt.legend()
plt.show()
````

- 120일선은 장기적 추세선 혹은 경기선이라고 불림
- 120일 동안의 종가를 대표하기 떄문에 지지와 저항의 세기가 강력하며, 만약 이런 지지 혹은 저항이 무너진 경우 장기적인 관점에서 추세가 반전되는 신호로 여겨짐
- 장기적인 관점에서 120일 선을 상향 돌파할 때 매수, 하향 돌파할 때 매도하는 전략을 취한다면 펀더멘탈이 괜찮은 종목이라면 긍정적인 투자 결과로 이어질 가능성이 높음

## 전략

- 기본적으로 이동평균선은 지지선과 저항선의 역할을 한다고 했으므로, 가장 쉽게 생각해볼 수 있는 전략은 지지선과 저항선을 이용한 매매
- 가격이 이동평균선보다 위에 형성되어 있는데 이동평균선이 있는 곳만큼 떨어지고 있다면 이동평균선이 지지선 역할을 할 가능성이 있으므로,이 근처에서 매수 포지션을 취한다.
- 반대로, 가격이 이동평균선보다 아래에 형성되어 있는데, 이동평균이 있는 곳만큼 상승하고 있다면, 이동평균선이 저항선의 역할을 할 가능성이 있으므로, 이 근처에서 매도 포지션

#### 1) 크로스 분석

- 매우 단순
- 여러 개의 이동평균선을 한 차트에 추가하여 단기선이 장기선(골든 크로스)을 상향 돌파하거나, 단기선을 장기선을 하향 돌파(데드 크로스)할 경우 매매 포지션
- 골든 크로스 -> 매수 포지션, 데드 크로스 -> 매도 포지션

````python
def cross_analysis(code, short_ma=60, long_ma=120, start_date="2018"):

    today = datetime.today().strftime('%Y-%m-%d')

    df = fdr.DataReader(code, start=start_date, end=today)
    
    df[f'MA{short_ma}']=df['Close'].rolling(short_ma).mean()
    df[f'MA{long_ma}']=df['Close'].rolling(long_ma).mean()

    df_p = df[[f"MA{short_ma}",f"MA{long_ma}", "Close"]]

    df_p = df_p.dropna()

    df_p['cross'] = df_p[f'MA{short_ma}'] < df_p[f'MA{long_ma}']
    df_p['cross'] = df_p['cross'].astype('int64')

    cross_points = []
    for i in range(2, len(df_p)):
        pre = df_p['cross'][i-1] 
        current = df_p['cross'][i]
        if pre != current:
            ma_long = df_p[f'MA{long_ma}'][i]
            ma_short = df_p[f'MA{short_ma}'][i]
            if ma_short > ma_long:
                cross_points.append((df_p.index[i].strftime("%Y-%m-%d"),"Golden Cross"))
            else:
                cross_points.append((df_p.index[i].strftime("%Y-%m-%d"), "Dead Cross"))
    plt.figure(figsize=(14, 10))
    plt.plot(df_p.index, df_p[f"MA{short_ma}"], label=f'{short_ma}-day',color='red')
    plt.plot(df_p.index, df_p[f"MA{long_ma}"], label=f'{long_ma}-day',color='yellow')
    plt.plot(df_p.index, df_p["Close"], label='close',color='silver')
    plt.legend()
    plt.show()
    
    return cross_points


    
````

````python
cross_analysis("005930", short_ma=60, long_ma=120, start_date="2022")
````

- 확실히 60 <-> 120 크로스가 주가의 상승과 하락을 잘 반영하는 것 같다.

````python
def ema_straight_arrangement(code, ma_list=[10, 20, 50], start_date="2018"):

    today = datetime.today().strftime('%Y-%m-%d')
    df = fdr.DataReader(code, start=start_date, end=today)[['Close']]
    for ma in ma_list:
        df[f'MA{ma}']=df['Close'].rolling(ma).mean()
    
    return df
````
